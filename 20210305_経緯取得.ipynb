```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm


def get_lat_lon_from_address(address_l):
    """
    address_lにlistの形で住所を入れてあげると、latlonsという入れ子上のリストで緯度経度のリストを返す関数。
    >>>>get_lat_lon_from_address(['東京都文京区本郷7-3-1','東京都文京区湯島３丁目３０−１'])
    [['35.712056', '139.762775'], ['35.707771', '139.768205']]
    """
    url = 'http://www.geocoding.jp/api/'
    latlons,lats,lons = [],[],[]
    err=[]
    for address in tqdm(address_l):
        payload = {"v": 1.1, 'q': address}
        r = requests.get(url, params=payload)
        ret = BeautifulSoup(r.content,'lxml')
        if ret.find('error'):
            latlons.append(["-","-"])
            lats.append("-")
            lons.append("-")
            err.append(address)
        #raise ValueError(f"Invalid address submitted. {address}")
        
        else:
            lat = ret.find('lat').string
            lon = ret.find('lng').string
            latlons.append([lat,lon])
            lats.append(lat)
            lons.append(lon)
            time.sleep(2)
    return latlons,lats,lons,err
```

```python
#Main
import pandas as pd
df = pd.read_excel('C:/Users/tong/Desktop/就活イベント一覧.xlsx',sheet_name="Sheet2")
data=list(df["住所"])
double,left,right,err=get_lat_lon_from_address(data)

df["緯度"]=left
df["経度"]=right
df.to_excel("C:/Users/tong/Desktop/就活イベント一覧（経緯取得）.xlsx", sheet_name='Sheet2',index=False)
df
```
